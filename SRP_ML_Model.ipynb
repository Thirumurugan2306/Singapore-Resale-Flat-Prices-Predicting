```python
import statistics

def get_median(x):
    split_list = x.split(' TO ')
    float_list = [float(i) for i in split_list]
    median = statistics.median(float_list)
    return median

df['storey_median'] = df['storey_range'].apply(lambda x: get_median(x))
df
```

```python
from sklearn.preprocessing import StandardScaler

X=df1[['cbd_dist','min_dist_mrt','floor_area_sqm','lease_remain_years','storey_median']]
y=df1['resale_price']

# Normalizing the encoded data
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

# test and train split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from joblib import dump
import numpy as np

def train_and_evaluate_random_forest(X_train, y_train, X_test, y_test):
    # Random Forest model with reduced parameters
    model = RandomForestRegressor(n_estimators=50, max_depth=10, max_features='sqrt', random_state=42)

    # Train the model
    model.fit(X_train, y_train)

    # Evaluate the model
    y_pred = model.predict(X_test)

    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)

    # Print evaluation metrics
    print("\nMean squared error:", mse)
    print('Mean Absolute Error:', mae)
    print('Root Mean squared error:', rmse)
    print("\nR-squared:", r2)

    return model, r2

# Train and evaluate Random Forest model with reduced parameters
random_forest_model, random_forest_r2 = train_and_evaluate_random_forest(X_train, y_train, X_test, y_test)

print(f"\nThe Random Forest model has R-squared: {random_forest_r2}")


```

```python
# Save the Random Forest model using joblib
dump(random_forest_model, 'random_forest_model_reduced.joblib')

print("Reduced Random Forest model saved as 'random_forest_model_reduced.joblib'")
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from joblib import dump
import numpy as np

def train_and_evaluate_model(model, X_train, y_train, X_test, y_test):
    # Train the model
    model.fit(X_train, y_train)

    # Evaluate the model
    y_pred = model.predict(X_test)

    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)

    # Print evaluation metrics
    print("\nMean squared error:", mse)
    print('Mean Absolute Error:', mae)
    print('Root Mean squared error:', rmse)
    print("\nR-squared:", r2)

    return model, r2

# Load your data (replace X_train, y_train, X_test, y_test with your actual data)
# X_train, y_train, X_test, y_test = load_your_data()

# Models to try
models_to_try = [
    RandomForestRegressor(n_estimators=50, random_state=42),
    DecisionTreeRegressor(max_depth=10, max_features='sqrt', random_state=42),
    GradientBoostingRegressor(n_estimators=50, learning_rate=0.1, max_depth=3, random_state=42)
]

best_model = None
best_r2 = -float('inf')

# Train and evaluate each model
for model in models_to_try:
    print(f"\nTraining and evaluating {type(model).__name__} model:")
    current_model, current_r2 = train_and_evaluate_model(model, X_train, y_train, X_test, y_test)

    # Save the best model based on R-squared
    if current_r2 > best_r2:
        best_model = current_model
        best_r2 = current_r2


```

```python
# Save the best model using joblib
if best_model is not None:
    dump(best_model, 'best_model.joblib')
    print("\nBest model saved as 'best_model.joblib' with R-squared:", best_r2)
else:
    print("\nNo best model found.")
```
